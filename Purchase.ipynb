## Read file

```python
import json

file = open("Record.json","r")
a = file.read()
record = json.loads(a)
file.close()
```

## Purchase

```python
print(' '*25, "WELCOME", ' '*20 ,"\n")
count = 1
j = []
i = 'Y'
while (i =='Y'):
    print("*"*13, "Please Enter Product Detail that You want to purchase" ,"*"*13)
    ui_prod  = input("\nEnter the product_Id: ")
    ui_quant = int(input("Enter the quantity: "))
    try:
        if(record[ui_prod]['qn'] < 0):
            print("\nSorry!! Item out of stock")
        elif(record[ui_prod]['qn'] < ui_quant):
            print("\nSorry!!. That much quantity is not available. \n")
        else:
            print("\n", "-"*17,"PRODUCT DETAILS :","-"*16, "\n" )          
            print("Product: ", record[ui_prod]['prod_name'])
            print("Original Price: ", record[ui_prod]['pr'])
            print("Avaible Dis: ", record[ui_prod]['discount'])
            a = int((record[ui_prod]['pr'] * record[ui_prod]['discount'])/100)
            print("Price after Discount : ", (record[ui_prod]['pr']-a))
            print( "Avaible quantity : ", record[ui_prod]['qn'])
            print("Billing Amount: ", record[ui_prod]['pr'] * ui_quant)
            print("Expiry date : ", record[ui_prod]['manu_date'])
            
            j.append({count : {'Prod_name':record[ui_prod]['prod_name'], 'Quantity' : ui_quant, 'Amount':(record[ui_prod]['pr']-a)* ui_quant}})
            
            count = count+1
            print("\nYour purchased detail : ", j)
            

            record[ui_prod]['qn'] = (record[ui_prod]['qn'] - int(ui_quant))
           

            print("\n", "*"*17,"THANKING YOU FOR VISITING","*"*16, "\n")
            print("*"*23,"VISIT AGAIN","*"*19, "\n")


    except KeyError:
        print("Please enter valid product id")

    a =input("Yes(Y)/No(N) = ")
    print("\n")
    if (a == 'N' or a == 'n'):
        print("Thanking you for visiting")
        break
    
        

 



```
