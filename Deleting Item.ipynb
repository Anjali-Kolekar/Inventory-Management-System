## Delete Item from Record

```python
file = open('Record.json')
a = file.read()
file.close()
```

```python
import json
record = json.loads(a)
```

```python
record
```

```python
print("*"*15, " DELETE ITEM ", "*"*15,"\n\n")

i1 = str(input("Please Enter Key to be Deleted : "))
if (i1 in record.keys()):
    del record[i1]
    print("\nDeleted Successfully")
    f = open('Record.json','w+')
    a = json.dumps(record)
    f.write(a)
    f.close()
else: 
    print("\nOops!!..Key not found")
    
```
